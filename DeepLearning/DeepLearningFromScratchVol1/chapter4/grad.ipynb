```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x[0]**2 + x[1]**2
```

```python
def numerical_gradient(f, x):
	h = 1e-4
	grad = np.zeros_like(x)

	for idx in range(x.size):
		tmp_val = x[idx]

		x[idx] = tmp_val+h
		fxh1 = f(x)

		x[idx] = tmp_val-h
		fxh2 = f(x)

		grad[idx] = (fxh1-fxh2)/(2*h)
		x[idx] = tmp_val

	return grad
```

```python
x0 = np.linspace(-2, 2, 10)
x1 = np.linspace(-2, 2, 10)
```

```python
X0, X1 = np.meshgrid(x0, x1)
```

```python
U = np.zeros_like(X0)
V = np.zeros_like(X1)

for i in range(X0.shape[0]):
    for j in range(X0.shape[1]):
        grad = numerical_gradient(f, np.array([X0[i, j], X1[i, j]]))
        U[i, j] = grad[0]
        V[i, j] = grad[1]

fig, ax = plt.subplots()
ax.quiver(X0, X1, -U, -V)

ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_aspect("equal")
ax.grid()

plt.show()
```

```python
def function_2(x):
    return x[0]**2 + x[1]**2
```

```python
def gradient_descent(f, init_x, lr=0.01, step_num=100):
	x = init_x

	for i in range(step_num):
		grad = numerical_gradient(f, x)
		x -= lr * grad
	return x
```

```python
init_x = np.array([-3.0, 4.0])
gradient_descent(function_2, init_x, lr=0.1)
```
