```python
def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=False)
    return x_test, t_test
```

```python
def init_network():
    with open("sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
    print(network.keys())
    return network
```

```python
def preidct(network, x):
    W1, W2, W3  = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = f.sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = f.sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = f.softmax(a3)
    return y
```

```python
x, t = get_data()
network = init_network()
```

```python
accuracy_cnt = 0
for i in range(len(x)):
    y = preidct(network, x[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1
print(accuracy_cnt / float(len(x)))
```
