```python
import torch

from multi_head_attention_wrapper import MultiHeadAttentionWrapper
```

```python
inputs = torch.tensor(
    [[0.43, 0.15, 0.89],
     [0.55, 0.87, 0.66],
     [0.57, 0.85, 0.64],
     [0.22, 0.58, 0.33],
     [0.77, 0.25, 0.10],
     [0.05, 0.80, 0.55]]
)
```

```python
d_in = inputs.shape[1] # 입력 임베딩 크기
d_out = 2 # 출력 임베딩 크기
batch = torch.stack((inputs, inputs), dim=0)
```

```python
torch.manual_seed(123)
context_length = batch.shape[1]
mha = MultiHeadAttentionWrapper(
    d_in, d_out, context_length, dropout=0.0, num_heads=2
)
context_vecs = mha(batch)
```

```python
print(context_vecs)
```

```python
print("context_vex.shape: ", context_vecs.shape)
```
