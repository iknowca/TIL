```python
import torch
```

```python
inputs = torch.tensor(
    [[0.43, 0.15, 0.89],
     [0.55, 0.87, 0.66],
     [0.57, 0.85, 0.64],
     [0.22, 0.58, 0.33],
     [0.77, 0.25, 0.10],
     [0.05, 0.80, 0.55]]
)
```

```python
query = inputs[1]
attn_scores_2 = torch.empty(inputs.shape[0])
for i, x_i in enumerate(inputs):
    attn_scores_2[i] = torch.dot(x_i, query)
print(attn_scores_2)
```

```python
attn_weights_2_tmp = attn_scores_2 / attn_scores_2.sum()
print("attention weights: ", attn_weights_2_tmp)
print("attention sum: ", attn_weights_2_tmp.sum())
```

```python
def softmax_naive(x):
	return torch.exp(x) / torch.exp(x).sum(dim=0)

attn_weights_2_naive = softmax_naive(attn_scores_2)
print("attention weights: ", attn_weights_2_naive)
print("attention sum: ", attn_weights_2_naive.sum())
```

```python
attn_weigths_2 = torch.softmax(attn_scores_2, dim=0)
print("attention weights: ", attn_weigths_2)
print("attention sum: ", attn_weigths_2.sum())
```

```python
query = inputs[1]
context_vec_2 = torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    context_vec_2 += attn_weigths_2[i]*x_i
print(context_vec_2)
```

```python
attn_scores = torch.empty(6, 6)
for i, x_i in enumerate(inputs):
    for j, x_j in enumerate(inputs):
        attn_scores[i, j] = torch.dot(x_i, x_j)
print(attn_scores)
```

```python
attn_scores = inputs @ inputs.T
print(attn_scores)
```

```python
attn_weights = torch.softmax(attn_scores, dim=1)
print(attn_weights)
```

```python
all_context_vectors = attn_weights @ inputs
print(all_context_vectors)
```
